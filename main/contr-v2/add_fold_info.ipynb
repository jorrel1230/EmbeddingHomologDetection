```python
# Create a mapping from (FA, SF, SF-UNIID) in scop_df to CF
mapping = {}
for _, row in scop_df.iterrows():
    key = (row['FA'], row['SF'], row['SF-UNIID'])
    mapping[key] = row['CF']

# Convert df columns to match the types in scop_df
df['fa_str'] = df['fa'].astype(str)
df['sf_str'] = df['sf'].astype(str)

# Add the fold column to df by looking up the corresponding CF value
df['fold'] = df.apply(lambda row: mapping.get((row['fa_str'], row['sf_str'], row['uid']), None), axis=1)

# Check how many rows have fold information
print(f"Rows with fold information: {df['fold'].notna().sum()} out of {len(df)}")

df = df[['index', 'uid', 'fa', 'sf', 'fold', 'seq']]
df
```
